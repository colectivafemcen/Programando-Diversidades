## Gráfico de superficie  

```python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y) # Entrega las coordenadas de una matriz desde un vector coordenado
Z = f(X, Y)
```

```python
fig = plt.figure(1, figsize = (12, 8))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, cmap=matplotlib.cm.coolwarm)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

# Animaciones 
Matplotlib no solo tiene la capacidad de crear gráficos, sino que además se pueden crear animaciones con este de manera muy simple. Para lo que necesitamos llamar la función animación de matplotlib:

```python
import matplotlib.animation as animation
```

```python
x = np.linspace(-1, 1, 100)
y = np.sin(x**2*25)
data = np.array([x, y])

fig = plt.figure()
line, = plt.plot([], [], "r-") # Comenzamos un plot que vacío
plt.axis([-1.1, 1.1, -1.1, 1.1])
plt.plot([-0.5, 0.5], [0, 0], "b-", [0, 0], [-0.5, 0.5], "b-", 0, 0, "ro")
plt.grid(True)
#plt.title("")

# Creamos una función que se llamará en cada iteración para actualizar la animación 
def update_line(num, data, line):
    line.set_data(data[..., :num] + np.random.rand(2, num) / 25)  # Se gráfica un número determinado de datos
    return line,

line_ani = animation.FuncAnimation(fig, update_line, frames=50, fargs=(data, line), interval=100)
plt.close() # Se cierra el plot para evitar que se cree un gráfico estático 
```

Ahora, debemos mostrar en pantalla nuestra animación. Para esto usaremos 2 métodos diferentes: \\
El primero consiste en usar HTML para desplegar la animación


```python
from IPython.display import HTML

HTML(line_ani.to_html5_video())
```

La otra manera simula los gráficos interactivos de JavaScript.

```python
HTML(line_ani.to_jshtml())
```
