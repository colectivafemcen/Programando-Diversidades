###Definicin de funciones

```python
def area_cuadrado(L):
  return L*L
```

```python
area_cuadrado(2)
```

```python
def linea_recta(x):
  return 3*x
```

```python
input = np.arange(0,5.1,0.1)
output = linea_recta(input)

```

```python
plt.plot(input,output)
plt.show()
```

```python
plt.plot(input,output,color='y') 
plt.ylabel('y') 
plt.xlabel('x')
plt.show()
```

```python
plt.plot(input,output,color='y')
plt.ylabel('y')
plt.xlabel('x')
plt.grid() 
plt.show()
```
